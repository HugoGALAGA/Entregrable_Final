```python
from avlw import AVLTree
import time
import sys  # Import the sys module

# --- Time function provided ---
def time_function(func, *args, **kwargs):
    start_time = time.perf_counter()
    result = func(*args, **kwargs)
    end_time = time.perf_counter()
    duration = end_time - start_time
    print(f'Execution time: {duration:.10f} secs') # Keep print for individual search times if needed during testing
    return duration

# Instantiate a new AVLTree object for search timing (40k initial nodes)
tree_avl_worst_search_40k = AVLTree()
num_nodes_initial_avl_worst_case_search_40k = 40000 # Increased to 40k initial nodes for AVL search test
num_searches_avl_worst_case = 25
start_value_searches_avl_worst_case = num_nodes_initial_avl_worst_case_search_40k + 1
end_value_searches_avl_worst_case = start_value_searches_avl_worst_case + num_searches_avl_worst_case - 1
search_values_avl_worst_case = range(start_value_searches_avl_worst_case, end_value_searches_avl_worst_case + 1)

# Increase recursion limit (be cautious with very large values) - likely not needed for AVL, but kept for consistency
recursion_limit = 41000  # Increased recursion limit to 41k (same as BST 40k)
sys.setrecursionlimit(recursion_limit)

# --- Insert initial 40000 nodes (worst-case order) into AVL Tree and measure insertion time ---
start_time_initial_avl_insert_search_40k = time.perf_counter()
tree_avl_worst_search_40k.insert_worst_case(num_nodes_initial_avl_worst_case_search_40k)
end_time_initial_avl_insert_search_40k = time.perf_counter()
insertion_time_initial_avl_worst_case_search_40k = end_time_initial_avl_insert_search_40k - start_time_initial_avl_insert_search_40k

print(f"Time to insert initial {num_nodes_initial_avl_worst_case_search_40k} nodes in worst-case order AVL Tree (for 40k search test): {insertion_time_initial_avl_worst_case_search_40k:.6f} seconds")

# --- Perform searches for the next 25 worst-case values in AVL Tree and measure individual search times ---
search_times_last_25_avl_40k = [] # List to store individual search times for AVL Tree (40k initial)
total_search_time_avl_worst_case_40k = 0
print("--- Timing individual searches for the next 25 worst-case values in AVL Tree (after 40k insertion) ---")
for search_value in search_values_avl_worst_case:
    search_time = time_function(tree_avl_worst_search_40k.search, search_value) # Time each individual search in AVL
    search_times_last_25_avl_40k.append(search_time) # Store individual search time
    total_search_time_avl_worst_case_40k += search_time # Accumulate AVL search times

print("\n--- End of individual search timings in AVL Tree (after 40k insertion) ---\n")

# --- Calculate the average search time for the last 25 searches in AVL Tree (40k initial) ---
average_search_time_last_25_avl_40k = sum(search_times_last_25_avl_40k) / len(search_times_last_25_avl_40k)

# --- Print the results for AVL Tree Search Test (40k initial) ---
print("--- Results for AVL Tree Search Test (Worst-Case Order, 40k initial nodes) ---")
print(f"Time to Insert Initial {num_nodes_initial_avl_worst_case_search_40k} nodes (Worst-Case Order AVL Tree): {insertion_time_initial_avl_worst_case_search_40k:.6f} seconds")
print(f"Total Time to Search for the next {num_searches_avl_worst_case} values (AVL Tree, 40k initial): {total_search_time_avl_worst_case_40k:.6f} seconds")
print(f"Total Time (Insertion + Search) for AVL Tree (Worst-Case Order, 40k initial): {insertion_time_initial_avl_worst_case_search_40k + total_search_time_avl_worst_case_40k:.6f} seconds") # Calculate total time directly here
print(f"\nList of search times (seconds) for the last 25 searches (AVL Tree, 40k initial):") # Add list of search times
print(search_times_last_25_avl_40k)
print(f"\nAverage Search Time for the last 25 searches (AVL Tree, 40k initial): {average_search_time_last_25_avl_40k:.10f} seconds") # Print average search time
print("--- End of Results (AVL Tree Search Test, 40k initial) ---\n")

# --- Store the average AVL search time for 40k initial nodes ---
avl_40k_avg_search_time = average_search_time_last_25_avl_40k # Store average search time in a new variable
```
